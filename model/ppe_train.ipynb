```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="construction-ppe.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    project="ppe_training",
    name="yolov8n_construction_ppe",
    exist_ok=True
)
```

```python
model = YOLO("/content/runs/detect/ppe_training/yolov8n_construction_ppe/weights/best.pt")

metrics = model.val(
    data="construction-ppe.yaml",
    imgsz=640,
    device=0
)
```

```python
model.predict(
    source="/content/datasets/construction-ppe/images/test",
    imgsz=640,
    conf=0.35,
    save=True
)
```
